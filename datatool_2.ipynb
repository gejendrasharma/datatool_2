```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
data = []

# Create multiple numerical values for each category
for category in categories:
    # Generate 100 random values for each category from a normal distribution
    values = np.random.normal(loc=np.random.randint(20, 80), scale=10, size=100)
    for value in values:
        data.append({'Category': category, 'Value': value})

# Create a DataFrame
df = pd.DataFrame(data)

# Create a box plot to visualize the distribution of the numerical variable across categories
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Value', palette='Set2')

# Add title and labels
plt.title('Distribution of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Show grid
plt.grid(axis='y')

# Show the plot
plt.show()
```

Q5>> Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
feature_1 = np.random.normal(loc=0, scale=1, size=n_samples)
feature_2 = feature_1 * 0.5 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # Correlated with feature_1
feature_3 = feature_1 * -0.3 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # Correlated with feature_1
feature_4 = np.random.normal(loc=0, scale=1, size=n_samples)  # Uncorrelated feature

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Step 1: Generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Step 2: Create a pivot table for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Step 3: Create a heatmap
fig_heatmap = px.imshow(pivot_table, 
                         title='Sales Variation Across Different Months and Days',
                         labels=dict(x='Month', y='Day', color='Sales'),
                         x=pivot_table.columns,
                         y=pivot_table.index)

# Show the heatmap
fig_heatmap.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Create x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

# Update layout
fig.update_layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X axis',
        yaxis_title='Y axis',
        zaxis_title='Z axis'
    )
)

# Show the plot
fig.show()
```

Q2>>Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 100, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame for better handling
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh scatter plot
output_notebook()  # Use this if you are in a Jupyter notebook; otherwise, use output_file("scatter_plot.html")
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add a scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']â€¨

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Prepare the data for Bokeh
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh bar chart
output_notebook()  # Use this if you are in a Jupyter notebook; otherwise, use output_file("bar_chart.html")
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)

# Add bars to the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits", fill_color="blue")

# Customize the legend
p.legend.title = "Fruits"
p.legend.location = "top_left"

# Show the plot
show(p)
```

Q4>> Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)â€¨

hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Generate random data
data_hist = np.random.randn(1000)

# Step 2: Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Prepare the data for Bokeh
# The edges need to be converted to the center of the bins for plotting
bin_centers = 0.5 * (edges[:-1] + edges[1:])
source = ColumnDataSource(data=dict(bin_centers=bin_centers, hist=hist))

# Step 3: Create a Bokeh plot
output_notebook()  # Use this if you are in a Jupyter notebook; otherwise, use output_file("histogram.html")
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add bars to the plot
p.vbar(x='bin_centers', top='hist', width=0.9, source=source, fill_color="blue", line_color="black")

# Show the plot
show(p)
```

Q5>>Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)â€¨

x = np.linspace(0, 1, 10)â€¨

y = np.linspace(0, 1, 10)â€¨

xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

# Step 1: Generate random data
data_heatmap = np.random.rand(10, 10)

# Step 2: Prepare the data
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Step 3: Create a Bokeh heatmap
output_notebook()  # Use this if you are in a Jupyter notebook; otherwise, use output_file("heatmap.html")

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())

# Create a figure
p = figure(title="Heatmap of Random Data", x_axis_label='X', y_axis_label='Y', width=600, height=600)

# Add rectangles to the plot
p.rect(x=x_flat, y=y_flat, width=0.1, height=0.1, source={'x': x_flat, 'y': y_flat, 'value': data_flat},
        fill_color=linear_cmap('value', palette="Viridis256", low=data_flat.min(), high=data_flat.max()),
        line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
